# Check Emojis

Count and print the count of emojis in a file

```python
import pandas as pd
from pathlib import Path
```

```python
data_path = Path(
    '../datasets/emotion_from_query/1149033514022649856-1148812793153499142_:anxious_face_with_sweat:.csv'
).resolve()
```

```python
data = pd.read_csv(data_path)
```

```python
import re
from emoji import demojize, emojize
```

```python
data['text'] = data.text.apply(lambda x: re.findall(r':[a-z_]+:', demojize(x)))
```

```python
emoji_dict = {}
for i, emojis in data.text.iteritems():
    for emoji in emojis:
        if emoji in emoji_dict:
            emoji_dict[emoji] += 1
        else:
            emoji_dict[emoji] = 1
```

```python
for emoji, count in sorted(emoji_dict.items(), key=lambda x: x[1], reverse=True):
    print(emojize(emoji) + '(' + emoji + '): ' + str(count))
```
